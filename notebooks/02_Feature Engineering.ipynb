## Constants

```python
TRAIN_VALID = '../data/raw/TrainAndValid.csv'
EXPORT_PATH = '../data/processed/TrainValid_proc.csv'
```

## Import tools

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

## Load the data

```python
df = pd.read_csv(TRAIN_VALID)
df.head()
```

```python
# Take a copy of original dataframe
df_copy = df.copy()
```

## Adding new date features

```python
df_copy['saleYear'] = pd.DatetimeIndex(df_copy.saledate).year
df_copy['saleMonth'] = pd.DatetimeIndex(df_copy.saledate).month
df_copy['saleDay'] = pd.DatetimeIndex(df_copy.saledate).day
df_copy['saleDayOfWeek'] = pd.DatetimeIndex(df_copy.saledate).dayofweek
df_copy['saleDayOfYear'] = pd.DatetimeIndex(df_copy.saledate).dayofyear
df_copy.drop('saledate', axis=1, inplace=True)
```

## Convert string columns to categorical

```python
for col in df_copy.columns:
    if pd.api.types.is_string_dtype(df_copy[col]):
      df_copy[col] = df_copy[col].astype('category')
```

```python
df_copy.info()
```

```python
df_copy.to_csv(EXPORT_PATH, index=False)
```
